# Data Description

# Prerequisites

## Import Library

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import gaussian_kde
import matplotlib.ticker as ticker
```

## tables

```python
skus = pd.read_csv('../../JD_sku_data.csv')
users = pd.read_csv('../../JD_user_data.csv')
clicks = pd.read_csv('../../JD_click_data.csv')
orders = pd.read_csv('../../JD_order_data.csv')
delivery = pd.read_csv('../../JD_delivery_data.csv')
inventory = pd.read_csv('../../JD_inventory_data.csv')
network = pd.read_csv('../../JD_network_data.csv')
```

## Data Preparation

```python
skus = pd.DataFrame(skus)
users = pd.DataFrame(users)
clicks = pd.DataFrame(clicks)
orders = pd.DataFrame(orders)
delivery = pd.DataFrame(delivery)
inventory = pd.DataFrame(inventory)
network = pd.DataFrame(network)
```

# SKUs table

```python
skus_att1 = skus.groupby(['attribute1']).size().reset_index(name='count')
skus_att1["percentage of SKUs"] = (skus_att1["count"]/skus_att1["count"].sum())*100
sns.barplot(data=skus_att1, x='attribute1', y='percentage of SKUs', color = 'b')
```

Chart showing the distribution of skus for attribute2 as a percentage

```python
skus_att2 = skus.groupby(['attribute2']).size().reset_index(name='count')
skus_att2 = skus_att2.reindex([0,2,3,4,5,6,7,8,1])
skus_att2["percentage of SKUs"] = (skus_att2["count"]/skus_att2["count"].sum())*100
sns.barplot(data=skus_att2, x='attribute2', y='percentage of SKUs', color = 'b')
```

# Clicks table

## Distribution of all click events across different channels

```python
clicks_channel = clicks.groupby(['channel']).size().reset_index(name='count')
clicks_channel["Percentage of clicks"] = (clicks_channel["count"]/clicks_channel["count"].sum())*100
sns.barplot(data=clicks_channel, x='channel', y='Percentage of clicks', color = 'b')
```

# Orders table

## Distribution of promise delivery time

```python
orders_promise = orders.groupby(['promise']).size().reset_index(name='count')
orders_promise["Percentage of orders"] = (orders_promise["count"]/orders_promise["count"].sum())*100
sns.barplot(data=orders_promise, x='promise', y='Percentage of orders', color = 'b')
```

## Sales in quantity by date and order type

```python
orders_date_type = sns.histplot(data=orders, x="order_date", hue="type", multiple="stack")
orders_date_type.xaxis.set_major_locator(ticker.MultipleLocator(5))
plt.xticks(rotation=45)
```

# Network table

## Number of districts within each region

```python
network_region = network.groupby(['region_ID']).size().reset_index(name='number of districts')
sns.barplot(data=network_region, x='region_ID', y='number of districts', color = 'b')
```
